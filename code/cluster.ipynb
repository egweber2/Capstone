```python
import numpy as np
import pandas as pd
```

```python
restaurant_pears = pd.read_csv('restaurant_pears.csv')
```

```python
restaurant_pears
```

```python
pivot = restaurant_pears.pivot('biz1','biz2','Correlation').fillna(0)
```

```python
pivot
```

```python
piv_arr = pivot.as_matrix()
```

```python
piv_arr
```

```python
np.fill_diagonal(piv_arr,1)
```

```python
# create distance
ones = np.ones((1000,1000))
restaurant_dist = (-1.0*(piv_arr-ones))/2.0
```

```python
restaurant_dist
```

```python
restaurant_dist_matrix = np.matrix(restaurant_dist)
```

# Fastcluster

```python
# convert to dissimilarity: 1-abs(correlation)
restaurant_dissimi = ones-abs(piv_arr)
```

```python
restaurant_dissimi
```

```python
%timeit 
data_link = linkage(restaurant_dissimi, method='single')#,preserve_input=True)
plt.figure(figsize=(28,20))
dendrogram(data_link)#,labels=data.dtype.names)
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.suptitle('Samples clustering', fontweight='bold', fontsize=14)
plt.show()
```
