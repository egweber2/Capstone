# Convert correlation to dissimilarity

```python
restaurant_pears = pd.read_csv('restaurant_pears.csv',index_col = 0)
```

```python
restaurant_pears
```

```python
pivot = restaurant_pears.pivot('biz1','biz2','Correlation').fillna(0)
```

```python
pivot
```

```python
piv_arr = pivot.as_matrix()
```

```python
piv_arr
```

```python
# fill diagonal as 1
np.fill_diagonal(piv_arr,1)
```

```python
# create distance
ones = np.ones((1000,1000))
```

```python
# convert to dissimilarity: 1-abs(correlation)
restaurant_dissimi = ones-abs(piv_arr)
```

```python
restaurant_dissimi
```

```python
restaurant_dissimi.shape
```

# Hiearchical Clustering

```python
%timeit 
data_link = linkage(restaurant_dissimi, method='single')#,preserve_input=True)
plt.figure(figsize=(28,20))
dendrogram(data_link,
          leaf_rotation=90.,  # rotates the x axis labels
          leaf_font_size=8.)#,labels=data.dtype.names)
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.suptitle('Samples clustering', fontweight='bold', fontsize=14)
plt.show()
```

```python
def fancy_dendrogram(*args, **kwargs):
    max_d = kwargs.pop('max_d', None)
    if max_d and 'color_threshold' not in kwargs:
        kwargs['color_threshold'] = max_d
    annotate_above = kwargs.pop('annotate_above', 0)

    ddata = dendrogram(*args, **kwargs)

    if not kwargs.get('no_plot', False):
        plt.title('Hierarchical Clustering Dendrogram (truncated)')
        plt.xlabel('sample index or (cluster size)')
        plt.ylabel('distance')
        for i, d, c in zip(ddata['icoord'], ddata['dcoord'], ddata['color_list']):
            x = 0.5 * sum(i[1:3])
            y = d[1]
            if y > annotate_above:
                plt.plot(x, y, 'o', c=c)
                plt.annotate("%.3g" % y, (x, y), xytext=(0, -5),
                             textcoords='offset points',
                             va='top', ha='center')
        if max_d:
            plt.axhline(y=max_d, c='k')
    return ddata
```

```python
last = restaurant_dissimi[-10:, 2]
last_rev = last[::-1]
idxs = np.arange(1, len(last) + 1)
plt.plot(idxs, last_rev)

acceleration = np.diff(last, 2)  # 2nd derivative of the distances
acceleration_rev = acceleration[::-1]
plt.plot(idxs[:-2] + 1, acceleration_rev)
plt.show()
k = acceleration_rev.argmax() + 2  # if idx 0 is the max of this we want 2 clusters
print "clusters:", k
```

```python
# set cut-off to 50
max_d = 50  # max_d as in max_distance
```

```python
#Truncate branches
%timeit 
data_link = linkage(restaurant_dissimi, method='ward')#,preserve_input=True)
plt.figure(figsize=(28,20))
fancy_dendrogram(data_link,
          truncate_mode='lastp',
          p=50,
          show_leaf_counts=False,  # otherwise numbers in brackets are counts
    leaf_rotation=90.,
    leaf_font_size=12.,
    show_contracted=True,# to get a distribution impression in truncated branches
    annotate_above=10,
    max_d=max_d
)#,labels=data.dtype.names)
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.suptitle('Samples clustering', fontweight='bold', fontsize=14)
plt.show()
```
